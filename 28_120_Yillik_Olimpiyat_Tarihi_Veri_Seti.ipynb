```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#python daki uyarıları kapatalım
import warnings
warnings.filterwarnings("ignore")

```

```python
veri=pd.read_csv("olimpiyatlar_temizlenmis.csv")
veri.head()
```

```python
#Saçılım grafiği çizmek için bir plot yapalım
def sacilimGrafik():
    sns.scatterplot(x='boy', y='kilo', data=veri)
    plt.title('Boy ve Kilo Dağılımı - Beyaz Izgara Tema')
    plt.show()

sacilimGrafik()
```

```python
#seaborn stilleri
sns.set_style("white")
sacilimGrafik()
```

```python
#seaborn stilleri
sns.set_style("darkgrid")
sacilimGrafik()
```

```python
#seaborn stilleri
sns.set_style("whitegrid")
sacilimGrafik()
```

```python
#hue=sınıf ya da etiket olarak adlandırılabilir
#scatter plot çizdirilirken hue'ya göre sınıflar farklı renkendirilebilir
sns.scatterplot(x='boy', y='kilo', hue='madalya', data=veri)
plt.title('Madalyaya Göre Boy ve Kilo Dağılımı')
plt.show()
```

Saçılım Grafiği - Doğrusal Regresyon

```python
#saçılım grafiği çizilir; üzerine doğrsual regresyon yöntemi ile line fit işlemi gerçekleştirilir
sns.regplot(x='boy', y='kilo', data=veri, marker='+', scatter_kws={'alpha':0.2})
plt.title('Boy ve Kilo Değişimi')
plt.show()
```

Renk Paletleri

Renk plaetleri görselleştirmede rengin gücünü kullanmak için belirlenmiş, içerisinde farklı renkleri barındıran yapılardır.

```python
#Renk paletleri
sns.scatterplot(x='boy', y='kilo', hue='madalya', data=veri, palette='Set1')
plt.title('Madalyaya Göre Boy ve Kilo Dağılımı')
plt.show()
```

```python
#Renk paletleri
sns.scatterplot(x='boy', y='kilo', hue='madalya', data=veri, palette='rocket')
plt.title('Madalyaya Göre Boy ve Kilo Dağılımı')
plt.show()
```
